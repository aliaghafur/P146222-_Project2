# **Introduction**

This dataset contains information about hotel bookings for two hotel groups from July, 2015 until August, 2017.  This dataset contains 3024 observations (rows) and 32 variables. This dataset has been taken from github, known as Hotel Data Analysis and has been written by Shridhar Kamathe. The datasets are from three csv files (Data_2015, Data_2016, Data_2017).

- Github Link: https://github.com/shridharkamathe/Hotel-Data-Analysis/tree/main/Hotel%20Data%20Analysis

By analysing this dataset, we can provide valuable insights to assist
and guide related businessess to improve the business, increase their profitability and optimize their business operations. Average daily rate (ADR) is one of the most important for hotel analysis , which can help the hotel increase their revenue. The higher the ADR means that more revenue per occupied room significantly boosts the hotel's overall profitability.

In this dataset contains two types of hotel: City Hotels and Resort Hotels. In my opinion,  these two types differ significantly in terms of their location, primary focus of business and availables amenities. For example, City Hotels typically located in urban centers, close to shopping areas and easily accessible by public transport. In contrast, Resort Hotels are often situated in scenic locations, such as near to beaches, mountains, or lakes.



### Variables Discriptions:
The followings are variables in dataset:

1. **hotel** : type of hotel
2. **is_canceled** : value indicating if the booking is canceled or not (0==no, 1==yes)
3. **lead_time** : no. of days before actual arrival of the guests
4. **arrival_date_year** : year of arrrival date
5. **arrival_date_month** : month of arrival date
6. **arrival_date_week_number** : week no. of year for arrival date
7. **arrival_date_day_of_month** : day of arrival date
8. **stays_in_weekend_nights** : no. of nights (Saturday and Sunday) guest booked
9. **stays_in_week_nights** : no. of nights (Mondat to Fiday) guest booked
10. **adults**: no. of adults
11. **children** : no of children
12. **babies** : no of babies
13. **meal**: Type of meal booked.  (Undefined/SC - no meal package);
(BB - Bed & Breakfast);
HB-Half board: (breakfast and one other meal- usually dinner);
(FB - Full board (breakfast, lunch and dinner)
14. **country** : country of origin (ISO 3155 - 3:2013)
15. **market_segment** : Online TA, Direct, Groups, Offline TA/TO, Complementary, Corporate, Aviation.
(TA-Travel Agents, TO-Tour Operators)
16. **distribution_channel** : TA/TO, Direct, Corporate, GDS.
17. **is_repeated_guest** : (0==no, 1==yes)
18. **previous_cancellations** : no. of previous bookings that were cancelled
19. **previous_bookings_not_canceled** : no of previous bookings that were not cancelled
20. **reserved_room_type**: Code of room type reserved (8 types)
21. **assigned_room_type** : Code for the type of room assigned to the guests (10 types)
22. **booking_changes** : no of changes made to the booking
23. **deposit_type** : No Deposit, Non Refund, Refundable
24. **agent** : ID of travel agency that made the booking
25. **company** : ID of company
26. **days_in_waiting_list** : no. of days the booking was in the waiting list before it was confirmed to the customer
27. **customer_type** : Transient, Transient-Party, Contrac, Group.
28. **adr** : Average daily rate
29. **required_car_parking_spaces** : no. of parking spaces required by the customer
30. **total_of_special_requests** : no. of special requests made by the customer
31. **reservation_status** : Canceled, Check-Out, No-Show
32. **reservation_starus_date** : date at which the last reservation status was updated.

```python
# combine the dataset from different csv file
df_2015 = pd.read_csv('/content/Data_2015.csv')
df_2016 = pd.read_csv('/content/Data_2016.csv')
df_2017 = pd.read_csv('/content/Data_2017.csv')

# concatenate the data frames
merged_df = pd.concat([df_2015, df_2016, df_2017], ignore_index=True)

# print the merged data
print(merged_df)
```

```python
# to get summary (data type and null values) from dataset using info
merged_df.info()

# covert data type from object to category:
categorical = ['hotel', 'meal', 'country', 'market_segment',
               'distribution_channel', 'reserved_room_type',
               'assigned_room_type', 'deposit_type',
               'customer_type', 'reservation_status']
for col in categorical:
  merged_df[col] = merged_df[col].astype('category')

print('\n-------------------------------------------------------------------\n')
# to verify:
merged_df.info()
```

The 'hotel' variable, intended to be a categorical variable with the values like 'City Hotel' and 'Resort Hotel', is currently represented as an object type. All other columns that represent categorical data are also type object. Object data type are general-purpose data type for strings, but we can change the type of data to category to make the analysis more efficient and clear.

```python
# change to CategoricalDtype - specific order (Jan, Feb, ..., Dec)
month_order= ['January', 'February', 'March', 'April', 'May', 'June',
         'July', 'August', 'September', 'October', 'November', 'December']
month_dtype = pd.CategoricalDtype(categories=month_order, ordered=True)
merged_df['arrival_date_month'] = merged_df['arrival_date_month'].astype(month_dtype)

# create new column 'booking_count' - how many booking for evey month
bookings_df = merged_df.groupby(['arrival_date_month', 'hotel'], observed=True)['hotel'].count().reset_index(name='bookings_count')

# line plot for the City Hotel
plt.figure(figsize=(12, 6))
sns.lineplot(x='arrival_date_month', y='bookings_count',
             data=bookings_df[bookings_df['hotel'] == 'City Hotel'],
             marker='o', label='City Hotel')

# line plot for the Resort Hotel
sns.lineplot(x='arrival_date_month', y='bookings_count',
             data=bookings_df[bookings_df['hotel'] == 'Resort Hotel'],
             marker='s', label='Resort Hotel')

plt.title('Booking Trends by Month, Booking Type, and Hotel')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.legend(title='Hotel and Booking Type', loc='lower right')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- The graph illustrate the number of bookings for City Hotels across different months of the year. Both hotels types experience varying levels of demand through the year, with August being a peak season for both. Usually July and August fall within summer months, where this months, school having summer break, so many families and individuals will use this time to go for vacation. Futhermore, favorable summer weather conditions often encorage many outdoor activities, further driving demand for both hotels.

- The lowest number of booking for both hotels are in January, this maybe because the peak holiday season just have passed like Christmas and New Year's and people might be recovering travel expenses or back to daily routines which go to school, and work. Beside that, after the excitment and often hectic nature of the holiday season, many people may simply desire a period of rest and relaxation at home, which this situation leading to decrease in travel demand.




```python
# change to CategoricalDtype - specific order (Jan, Feb, ..., Dec)
month_dtype = pd.CategoricalDtype(categories=month_order, ordered=True)
merged_df['arrival_date_month'] = merged_df['arrival_date_month'].astype(month_dtype)

# create new column adr - which is  average adr
adr_df = merged_df.groupby(['arrival_date_month', 'hotel'], observed=True)['adr'].mean().reset_index()

# line plot for the City Hotel
plt.figure(figsize=(12, 6))
sns.lineplot(x='arrival_date_month', y='adr',
             data=adr_df[adr_df['hotel'] == 'City Hotel'],
             marker='o', label='City Hotel')

# line plot for the Resort Hotel
sns.lineplot(x='arrival_date_month', y='adr',
             data=adr_df[adr_df['hotel'] == 'Resort Hotel'],
             marker='s', label='Resort Hotel')

plt.title('Average Daily Rate (ADR) by Month, Booking Type, and Hotel')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.legend(title='Hotel and Booking Type', loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- This graph ilustrate the ADR for both City Hotels and Resort Hotels across different months. City Hotels shows a generally upward trend from January to May, followed by a slighly decline towards the end of the year. In contrast, Resort Hotels shows a significant peak in ADR during August, with a subsequent decline in the following months.

- Although, for City Hotel has highest number of booking during summer month (July and August), but for ADR show that decline during that months. This can be attribute to several factors which during this month there is likely a surge in demand for hotel. This increased competition among hotel that can lead to hotel to offer discounts or promotions to attract guests. Beside that, guests often book hotels in advance, especially during peak season, and the hotel might offer earlier booking discount to incentive reservation which lead to a lower ADR.

```python
# plot for'stays_in_weekend_nights' and 'stays_in_week_nights' by 'hotel'
# group data by hotel type
# calculate average for night stay
hotel_stays = merged_df.groupby('hotel')[['stays_in_weekend_nights', 'stays_in_week_nights']].mean().reset_index()

# from 'hotel_stay', there are three column - 'hotel', 'stays_in_weekend_nights', 'stays_in_week_nights'
# add another column (average_night)
# 'stays_in_weekend_nights', 'stays_in_week_nights' become one column = 'stay_type'
hotel_melt = hotel_stays.melt(id_vars=['hotel'], var_name='stay_type', value_name='average_nights')

# create the grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='hotel', y='average_nights', hue='stay_type', data=hotel_melt, palette='muted')
plt.title('Average Stays in Weekend and Week Nights by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Average Number of Nights')
plt.legend(title='Stay Type')
plt.show()
```

 - Although many customers prefers to choose City Hotel, but in terms of time stay at the hotel, guests tend to spend longer durations at Resort Hotels. This is likely due to the fact that business travelers who often have shorter stays for meetings or conferences are more likely to stay at City Hotel. In the other hand, for Resort Hotels attract more leisure travelers who are looking for longer vacations and extended stays. Usually Resort Hotels often located near to beach and outside urban area, so there are many activities that guests can do such as beach volleyball, swimming, sunbathing, spas treatments and others. These will encourage guests to extend their stays and enjoy their leisure time.

- Furthermore, guests also tend to stay more on weekdays while the shorter average stay for weekend maybe because many guests use weekend for short gateways or just to take a time to relax their mind before start a busy workweeks. Another factor that may contribute to this result is the price per room difference between weekdays and weekends. Weekday rates are often lower compared to weekend rates, which can be significantly higher, potentially double from normal price.

```python
# 'assigned_room_type' vs '_is_canceled'
plt.figure(figsize=(12, 6))
sns.countplot(x='assigned_room_type', hue='is_canceled', data=merged_df, palette='Set2')
plt.title('Assigned Room Type vs. Cancellation')
plt.xlabel('Assigned Room Type')
plt.ylabel('Count')
plt.legend(title='Cancellation', labels= ['No', 'Yes'])
plt.show()
```

- As we look before, the most popular room choose by guests during resrvation is Room A, but as the plot bar 'Assigned Room Type vs. Cancellation' above show that Room A also has a lower cancellation.

- There maybe some factors that contribute to this observations. Firstly, Room A, with attractive features such as larger size, better view , and potentially affordable in price, may lead to higher guest satisfaction and contribute less likely to cancel. In addition, other than Room A, we can see that the number of non-cancelled booking is generally lower compared to the number of cancellation. This shows that guests satisfied with the assigned room. If the assigned room does not meet their expectation or if unexpected additional cost arise, guests may be more likely to cancel the reservations.


```python
# lead_time' vs 'is_canceled'
plt.figure(figsize=(10, 6))
sns.histplot(data=merged_df, x='lead_time', hue='is_canceled',
             palette='Set2', multiple='stack')
plt.title('Distribution of Lead Time by Cancellation Status')
plt.xlabel('Lead Time (days)')
plt.ylabel('Count')
plt.legend(title='Cancellation', labels= ['Yes', 'No'])
plt.show()
#
```

- The bar plot indicates that as the the numbers of days before actual date decrease, the number for guest cancellations generally decrease. However, there is still a chance that the guests will cancel the booking even when they are close to the actual date. The lower number of cancellation may be because of guests tend to carefully think before decide decision regarding their travel plan.
